```python
def generate_data(
    n_dims: int,
    n_samples: int,
    from_distribution: str="normal", # Distribucion de proba a partir de la cual se van a generar los datos
    distribution_params: Dict[str, float]={"mean": 0., "std": 1.} # Diccionario con dos elementos: un str y un float. En este caso, definidos por default
) -> np.array:
    """
    Descripción:
    Esta funcion genera datos artificiales muestreados de una distribución de probabilidad especificada.
    
    Entradas:
    - n_dims: Número de dimensiones (características) que contendrá cada ejemplo 
    - n_samples: Número de muestreos (ejemplos) que deberán ser generados
    - from_distribution: Distribución de la que deberán ser muestreados los puntos, "normal" o "uniform"   
    - distribution_params: Parámetros para la distribución de la que serán muestradas los valores.
        "uniform": {
            "low": El número mínimo que puede ser muestreado
            "high": El número máximo que puede ser muestreado
            }
        "normal": {
            "mean": La media que tendrá la distribución normal
            "std": La desviación estándar de la distribución normal
            }
            
    Salidas:
    - Un arreglo de numpys muestreado de la distribución seleccionada con dimensiones (n_samples, n_dims)
    """
 
    # CÓDIGO AQUÍ
    
    
    rng = np.random.default_rng()

    # Verificamos cuál es la distribucion de probabilidad solicitada por el usuario y generamos los datos de acuerdo a la distribucion correspondiente
    # Normal: Pasamos la media, la std y las dimensiones del numpy array a generar
    if from_distribution == "normal":
      examples = np.random.default_rng().normal(distribution_params["mean"], distribution_params["std"], (n_samples, n_dims))
    elif from_distribution == "uniform":
      examples = np.random.default_rng().uniform(distribution_params["low"],distribution_params["high"],(n_samples, n_dims))
    # Uniforme: Pasamos limite inferior (inclusive) y limite superior (exclusivo o menor que) más las dimensiones del numpy array a generar

    return examples
```

```python
__n_samples__ = 1000
__n_dims__ = 2

__uniform__ = generate_data(
    n_dims=__n_dims__, 
    n_samples=__n_samples__, 
    from_distribution="uniform", 
    distribution_params={"low": -10., "high": 10.}
)
# print(__uniform__)

__normal__ = generate_data(
    n_dims=__n_dims__, 
    n_samples=__n_samples__, 
    from_distribution="normal", 
    distribution_params={"mean": 3., "std": 2.}
)


assert __uniform__.shape == (__n_samples__, __n_dims__) # assert : Verifica que la condicion cumplida sea True, si no es así, arroja error
assert __normal__.shape == (__n_samples__, __n_dims__)

__fig__, __axs__ = plt.subplots(ncols=2, figsize=(15, 5)) # Generamos dos graficos en vertical
__axs__[0].set(title="Uniform data")
__axs__[1].set(title="Normal data")
_ = sns.scatterplot(x=__uniform__[:, 0], y=__uniform__[:, 1], ax=__axs__[0])
_ = sns.scatterplot(x=__normal__[:, 0], y=__normal__[:, 1], ax=__axs__[1])
```

```python
def label_data(
    data: np.array,
    model: np.array=None
) -> np.array: 
    """
    Descripción:
    Esta funcion genera el etiquetado de datos artificiales de acuerdo a un modelo generado. 
    En caso de no ser recibido como parámetro, el modelo debe ser generado aleatoriamente de 
    una distribución normal con media 0 y varianza 1.
    
    Entradas:
    - data: Datos a etiquetar de dimensiones Número de dimensiones (# ejemplos, # características)
    - model: Modelo utilizado para etiquetar
            
    Salidas:
    - Un arreglo de numpy referente a las etiquetas asignadas a cada ejemplo recibido en la entrada "data".
    - Un arreglo de numpy referente al modelo generado aleatoriamente.
    """


   ###################### CÓDIGO AQUÍ ########################################## 

    if model is None:
      # Generacion del modelo

      # Inicializamos el generador de numeros aleatorios
      rng = np.random.default_rng()

      #Calculamos el numero de dimensiones a calcular para el modelo
      dimensionesW = data.shape[len(data.shape)-1]

      # Generamos ponderaciones en un rango especifico
      model = []
    
      for indice in range(dimensionesW):
        model.append(round(rng.standard_normal(),2))

      # Definimos el bias
      bias  = 1

      # Agregamos el bias al modelo 
      model = np.array(model)
      model = np.insert(model, 0, bias)
  
    # Predicciones 

    # Creamos la columna de x_0 que corresponde al valor x del bias para ser agregado en el conjunto de datos
    x_0 = np.ones((len(data),), dtype=int)
    # print(x_0)

    # Agregamos la columna del valor de x_0 para el bias al principio del dataset
    data = np.insert(data, 0, x_0, axis=1)

    # Transponemos los vectores w y x (ambos deben ser transpuestos para realizar la multiplicacion)
    modelT = model.T

    prediccionesH = []
    for fila in data:
      h = np.dot(modelT, fila.T)
      h = np.sign(h)
      prediccionesH.append(h)      

    prediccionesH = np.array(prediccionesH)
    labels = prediccionesH

    return labels, model

```

```python
def compute_line_points(
    model: np.array,
    bounds: Tuple[float, float]=(-10, 10)
) -> np.array: 
    """
    Descripción:
    Esta funcion genera los puntos en los extremos de la linea generada por un modelo bidimensional 
    recibido como parámetro.
    
    Entradas:
    - model: El modelo que representa una linea. El modelo debe tener exactamente tres parámetros.
    - bounds: Indíca el límite inferior y superior del eje x, en un plano cartesiano, para los cuales 
            deben ser obtenidos los valores de y
            
    Salidas:
    - Un arreglo de numpy de tamaño 2x2 (una matriz), donde cada fila representa las coordenadas (x, y) de
        la linea
    """
    
    assert model.shape == (3,)    
    points = __random_fn__((2, 2)) 
    
   ###################### CÓDIGO AQUÍ ########################################## 
    
    indiceTupla = 0
    coordenada_x = 0
    coordenada_y = 0

    # Calculo despejando la variable target de mi modelo
    for x in range(points.shape[0]):
      y = 0
      for indice in range(model.shape[0]):
        # Si el indice es igual al indice del termino final del modelo
        if indice == (model.shape[0]-1):
          y = y / model[indice]
        elif indice == 0:
          y = -1*(model[indice])
        else:
          y = y - (1*(model[indice] * bounds[indiceTupla]))
          indiceTupla += 1 
      points[coordenada_x, coordenada_y] = bounds[coordenada_x]
      points[coordenada_x, coordenada_y+1] = y
      coordenada_x += 1
    
    return points
```

```python
def train_pla(
    data,
    target,
    max_iters=1000
) -> np.array: 
    """
    Descripción:
    Esta funcion ejecuta una variante del algoritmo PLA para clasificar datos linealmente separables.
    
    Entradas:
    - data: Indica las características utilizadas para la clasificación.
    - target: Indíca las etiquetas que se esperan obtener despues de una correcta clasificación.
    - max_iters: Indica el número máximo de iteraciones del algoritmo. Si el algoritmo no converge
                antes del número máximo de iteraciones, se detiene y devuelve los parámetros 
                calculados hasta el momento.
            
    Salidas:
    - Un arreglo de numpy indicando los parámetros calculados
    """
    # Se genera un array de tres elementos en este caso, que son los tres pesos que integrararn el modelo final w  
    model = __random_fn__((data.shape[1] + 1,))
    
   ###################### CÓDIGO AQUÍ ########################################## 
    
    # NOTA: x(t) es el vector de caracteristicas del registro t
    
    # Se genera un modelo aleatorio w
    predicciones, model = label_data(data)

    # PLA
    # Agregamos la columna de 1s al inicio del dataset
    # Creamos la columna de x_0 que corresponde al valor x del bias para ser agregado en el conjunto de datos
    x_0 = np.ones((len(data),), dtype=int)

    # Agregamos la columna del valor de x_0 para el bias al principio del dataset
    x = np.insert(data, 0, x_0, axis=1)

    # Se genera un numpy array de valores booleanos para conocer los valores predichos (target) que no son iguales a nuestro dataset original (y_dataset)
    precision = (predicciones == target)
    
    # Se recuperan los indices donde no se cumple la igualdad
    indicesIncorrectos = np.where(~precision) # Es una tupla donde el primer elemento de la tupla son los indices mal clasificados
    indicesIncorrectos = indicesIncorrectos[0] # Lo convertimos a un numpy array
    # print(indicesIncorrectos)

    numIndicesIncorrectos = len(indicesIncorrectos)

    iteraciones = 0
    while (numIndicesIncorrectos > 0) and (iteraciones <= max_iters):
          indice = indicesIncorrectos[0]
          model = model + (x[indice,:]*target[indice]) # DUDA: Este valor y(t)x(t) se suma solo al valor del bias o a todos los pesos de cada una de las caracteristicas del modelo
          iteraciones += 1
          predicciones, modelx = label_data(data, model)

          # Se genera un numpy array de valores booleanos para conocer los valores predichos (target) que no son iguales a nuestro dataset original (y_dataset)
          precision = (predicciones == target)
    
          # Se recuperan los indices de los arreglos donde no se cumple la igualdad
          indicesIncorrectos = np.where(~precision) # Es una tupla donde el primer elemento de la tupla son los indices mal clasificados
          indicesIncorrectos = indicesIncorrectos[0] # Lo convertimos a un numpy array
          numIndicesIncorrectos = len(indicesIncorrectos)
    
    return model
```
